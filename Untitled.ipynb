```python
import kagglehub
import os, shutil

# 1) lae alla (läheb kagglehub cache'i)
src_dir = kagglehub.dataset_download(
    "olaflundstrom/data-on-traffic-accidents-with-human-victims"
)
print("Cache kaust:", src_dir)

# 2) kopeeri kõik failid notebooki kausta
dst_dir = os.getcwd()
for fname in os.listdir(src_dir):
    src = os.path.join(src_dir, fname)
    dst = os.path.join(dst_dir, fname)
    if os.path.isdir(src):
        shutil.copytree(src, dst, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dst)

print("Kopeeritud notebooki kausta:", dst_dir)
```
